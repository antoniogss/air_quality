#Adding the column 'estacao'

```python
# assign dataset names
list_of_names = ['a_cp','b_cp','c_cp', #[0-1-2]:0
                 'a_ib','b_ib','c_ib', #[3-4-5]:1
                 'a_in','b_in','c_in', #[6-7-8]:2
                 'a_pd','b_pd','c_pd'] #[9-10-11]:3

# create empty list
df = []
  
# append datasets into teh list
for i in range(len(list_of_names)):
    temp_df = pd.read_csv("./"+list_of_names[i]+".csv",
                          sep = ';', 
                          index_col = None, 
                          header = 0,
                          error_bad_lines = False,
                          encoding='latin-1'
                        )
    df.append(temp_df)
```

```python
estacao = ["Capão Redondo", "Ibirapuera", "Interlagos", "Parque Dom Pedro II"]

df[0]['estacao'] = df[1]['estacao'] = df[2]['estacao'] = estacao[0]
df[3]['estacao'] = df[4]['estacao'] = df[5]['estacao'] = estacao[1]
df[6]['estacao'] = df[7]['estacao'] = df[8]['estacao'] = estacao[2]
df[9]['estacao'] = df[10]['estacao'] = df[11]['estacao'] = estacao[3]
```

```python
df[11].head()
```

#Concatenating files - 2

```python
df_a = pd.concat(
                  [df[0], df[3], df[6], df[9]], 
                  axis = 0, 
                  sort = False
                )

df_b = pd.concat(
                  [df[1], df[4], df[7], df[10]],
                  axis = 0, 
                  sort = False
                )

df_c = pd.concat(
                  [df[2], df[5], df[8], df[11]], 
                  axis = 0, 
                  sort = False
                )
```

```python
df_a.head()
```

```python
print('Columns: df_a ')
df_a.columns
```

```python
print('Columns: df_b ')
df_b.columns
```

```python
print('Columns: df_c ')
df_c.columns
```

```python
df_ab = pd.merge(df_a, #only two dfs at time
                 df_b, 
                 on=['Unnamed: 0', 'Unnamed: 1', 'estacao'],
                 how = "outer"
                )
```

```python
df = pd.merge(df_ab, 
              df_c, 
              on=['Unnamed: 0', 'Unnamed: 1', 'estacao'],
              how = "outer"
            )
```

```python
df.head()
```

#Renaming columns

```python
df.columns
```

```python
df.rename(
          columns = {
                      'Unnamed: 0':'data', 
                      'Unnamed: 1':'hora',
                      'O3(Ozônio) - µg/m3':'O3',
                      'RADG(Radiação Solar Global) - W/m2':'RADG',
                      'RADUV(Radiação Ultra-violeta) - W/m2':'RADUV',
                      'NO(Monóxido de Nitrogênio) - µg/m3':'NO',
                      'NO2(Dióxido de Nitrogênio) - µg/m3':'NO2',
                      'NOx(Óxidos de Nitrogênio) - ppb':'NOx',
                      'TEMP(Temperatura do Ar) - °C':'TEMP',
                      'UR(Umidade Relativa do Ar) - %':'UR',
                      'VV(Velocidade do Vento) - m/s':'VV'
                    },
          inplace = True
)

df.head()
```

#Changing commas to dot

```python
df = df.replace({',' : '.'}, regex = True)
```

#Converting string to float

```python
df.columns
```

```python
df['O3'] = pd.to_numeric(df['O3'], downcast='float', errors='coerce')
df['RADG'] = pd.to_numeric(df['RADG'], downcast='float', errors='coerce')
df['RADUV'] = pd.to_numeric(df['RADUV'], downcast='float', errors='coerce')
df['NO'] = pd.to_numeric(df['NO'], downcast='float', errors='coerce')
df['NO2'] = pd.to_numeric(df['NO2'], downcast='float', errors='coerce')
df['NOx'] = pd.to_numeric(df['NOx'], downcast='float', errors='coerce')
df['TEMP'] = pd.to_numeric(df['TEMP'], downcast='float', errors='coerce')
df['UR'] = pd.to_numeric(df['UR'], downcast='float', errors='coerce')
df['VV'] = pd.to_numeric(df['VV'], downcast='float', errors='coerce')
```

```python
df.describe()
```

#Formatting dates and times

```python
df['hora'].replace('24:00', '00:00', inplace = True)
gc.output.clear()
```

```python
df['data_hora'] = df['data'] + ' ' + df['hora']
df.head()
```

```python
df = df.drop(
              [
                'data', 
                'hora'
              ],
              axis = 1
)
df.head()
```

```python
df['data_hora'] = pd.to_datetime(df['data_hora'], format = '%d/%m/%Y %H:%M')
df.head()
```

```python
df.dtypes
```

#Adding the column  days of the week

```python
dias_ptbr = {
              'Sunday':'Domingo', 
              'Monday':'Segunda-feira', 
              'Tuesday':'Terça-feira', 
              'Wednesday':'Quarta-feira', 
              'Thursday':'Quinta-feira', 
              'Friday':'Sexta-feira', 
              'Saturday':'Sábado'
}

df['dia_da_semana'] = df['data_hora'].dt.day_name().replace(dias_ptbr)

df.head()
```
